```python
df_all_year= pd.DataFrame()
df_all_year=  pd.read_csv('marcap-2021.csv.gz', dtype={'Code':str}, parse_dates=['Date'])
```

```python
for files in forders:
    df= pd.read_csv(files, dtype={'Code':str}, parse_dates=['Date'])
    if df['Code'].isin(df_all_year['Code']):
    df_all_years = pd.concat([df_all_years, df])
```

```python
df_all_years= df_all_years.sort_values(by=['Code','Date'])
df_all_years
```

```python
df_all_years= df_all_years.reset_index()
df_all_years= df_all_years.drop(['index'], axis=1)
df_all_years
```
