````python
# df_new의 행과 열 갯수를 각각 row와 column 변수로 받음

(row,column) = df_new.shape
print(row)
print(column)
````

````python
def y_to_t_percent(i,j):
    while(i < row):
        df_new.iloc[i,0]= ''
        while(j < column):
            df_new.iloc[i,j]=(df.iloc[i,j]-df.iloc[i,j-1])/df.iloc[i,j-1]*100   
            j += 1
        i+= 1
        j=1
        
y_to_t_percent(0,1)
````

### y_to_t_percent 함수 설명

**어제부터 오늘의 주가 변동율   "y_to_t_percent"  
(y_to_t_percent) = (오늘 시가 - 전날 시가)/ 전날 시가 * 100**  
  
"종목코드"를 행 인덱스, "일자"를 열 인덱스로 "시가" 데이터만 저장한 데이터프레임 df에서  
df의 열끼리 연산하여 (y_to_t_percent)값을 df_new에 채우기  
  
df_new에서 첫번째 열(첫번째 날짜)의 "시가"는 비교할 대상(전날 시가)이 없기 때문에 계산에서 제외함   
df_new의 첫번째 열에는 값을 비워 둠(NaN)
  
  
#### while문  
행 인덱스(종목코드)는 i, 열 인덱스(일자)는 j 변수   
i는 row보다 작을 동안 while문 실행, j는 column보다 작을 동안 두번째 while문 실행  
*(인덱스 번호가 0부터 시작하기 때문에 (row-1), (column-1)까지가 행과 열 전체 인덱스임)*  

```python
    df_new.iloc[i,0]= ''
```
-> df_new의 첫번째 열(첫째날)에는 값을 비워 둠

```python
    while(j < column):
        df_new.iloc[i,j]=(df.iloc[i,j]-df.iloc[i,j-1])/df.iloc[i,j-1]*100   
        j += 1
```
1. j는 첫번째 열을 제외하고 두번째 열(index=1)부터 시작해서 while문 실행
2. df_new의 i=0행 j열의 값= ((오늘 시가)-(전날 시가)) / (전날 시가) * 100
(전날 시가 대비 오늘 시가가 얼마나 변했는지 계산해여 하므로 오늘 시가에서 전날 시가를 뺌)
3. "j+=1"을 통해 한칸 옆의 열로 이동해서 계산식 반복을 통해 첫번째 행(index=0)값 채우기

```python
while(i < row):
    df_new.iloc[i,0]= ''
    while(j < column):
        df_new.iloc[i,j]=(df.iloc[i,j-1]-df.iloc[i,j])/df.iloc[i,j-1]*100   
        j += 1
    i+= 1
    j=1
```
1. "i+=1"을 통해 한칸 아래 행으로 이동
2. "j=1"로 열 인덱스값 초기화
3. 다시 바깥쪽 while문의 처음으로 돌아가서 두번째 행(index=1)의 첫번째 열(index=0)의 값 비워둠
4. 현재 "j=1"이기 때문에 2번재 행의 두번째 열부터 안쪽 while문 다시 실행
5. 제일 마지막 행까지 중첩 while문을 실행하여 "df_new" 채움
