```python
# 월별 시가 총합을 담을 리스트 "sp_sum" 생성

sp_sum= []
type(sp_sum)
```

```python
# 월 시가 총합 "sp_sum" 리스트 채우기
# append 함수 이용해서 리스트 값 추가

sp_ind_f = 0
for d_ind in days:
    sp_ind_b = d_ind + sp_ind_f    # 누적 일수 + 다음 일수

    sp_sum.append(sum(startprice[sp_ind_f : sp_ind_b]))
    sp_ind_f = sp_ind_b
```

```python
len(sp_sum)
```

```python
# 월별 시가총액과 일수 리스트를 "df_monthly_avg"의 열로 추가

df_monthly_avg['월별 시가총액']=sp_sum
df_monthly_avg['일수']=days
df_monthly_avg
```

```python
# 월별 평균시가를 계산해서 새로운 column '평균시가'를 생성하고 값 저장함
# (월별 평균시가) = (월별 시가총합) / (일수)

df_monthly_avg['평균시가']= df_monthly_avg['월별 시가총액']/df_monthly_avg['일수']
df_monthly_avg
```

```python
# df_new의 행과 열 갯수를 각각 row와 column 변수로 받음
# monthly percent= (이번달 평균시가 - 지난달 평균시가)/ 지난달 평균시가 * 100

(row,column) = df_new.shape

def monthly_percent(i,j):
    while(i < row):
        df_new.iloc[i,0]= ''
        while(j < column):
            df_new.iloc[i,j]=(df.iloc[i,j]-df.iloc[i,j-1])/df.iloc[i,j-1]*100   
            j += 1
        i+= 1
        j=1
        
monthly_percent(0,1)
```

```python
# '일' column의 값에서 "'"와 ")" 제거

df_temp_split['일']=df_temp_split['일'].astype('str')
df_temp_split['일']=df_temp_split['일'].str[2:-2]
```

```python
df_temp_split
```

```python
# df_temp에 "df_temp_split" 합쳐서 'temp'열 분리한 열 생성

df_temp=pd.concat([df_temp,df_temp_split], axis=1)
df_temp
```
