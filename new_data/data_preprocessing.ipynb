```python
import pandas as pd
```

```python
df_all= pd.read_csv('./df_all_years.csv')
```

```python
df_all.head()
```

```python
df_all_needed= df_all[['Code','Name','Close','Open','High','Low','Date']]
df_all_needed.head()
```

```python
# 하루동안의 주가 변동율  "day_percent"
# (day_percent) = (종가 - 시가)/ (시가) * 100 
# day_percent를 df의 새로운 column으로 추가

df_all_needed['day_percent']= (df_all_needed['Close']-df_all_needed['Open'])/df_all_needed['Open']*100
```

```python
df_all_needed['Date']= df_all_needed['Date'].astype('str')
df_all_needed['Date'] = df_all_needed['Date'].str[:4]+df_all_needed['Date'].str[5:7]+df_all_needed['Date'].str[8:10]
df_all_needed
```

```python
df= df_all_needed[['Code','Date','Name']]
df= df.set_index(['Code','Date'])
df= df.unstack('Date')
df
```

```python
df= df.iloc[:, -10:].dropna()
df = df.stack('Date')
df
```

```python
df=df.reset_index()
df
```

```python
df_dropped= pd.merge(df, df_all_needed, how="left")
df_dropped
```

```python
df_dropped.to_csv('dropped_to_current.csv', index=False)
```
